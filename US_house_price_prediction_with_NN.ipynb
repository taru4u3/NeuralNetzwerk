Let us begin mit Explorative Data Analysis

```python
# Let us see the price distribution which is our variable to be predicted
plt.figure(figsize=(12,6))
sns.distplot(df.price)
```

```python
df.corr()['price'].sort_values(ascending=False) # correlation between price and other variables
```

```python
# Let us check the most coorelated feature sqftliving w.r.t price
sns.scatterplot(x='sqft_living',y='price',data=df)
```

```python
# Let us check the if there is a logical binding between bedrooms and price
sns.boxplot(x='bedrooms',y='price',data=df)
```

```python
sns.scatterplot(x='price',y='grade',data=df)
```

```python
# Let us see the relation between location and price so, between latitude and price
sns.scatterplot(x='long',y='price',data=df)
```

```python
# longitude and price
sns.scatterplot(x='lat',y='price',data=df)
```

Now let us put latitude and longitude both in perspective and see the connection with price

```python
sns.scatterplot(x='long',y='lat',data=df,hue='price')
```

The above visualization shows around top 1 % in certain lat and log have higher price but rest are all dimmed. So to visualise better the rest let us cut out the top 1 % that are already clear from above plot.

```python
df.sort_values('price',ascending=False).head(20)
```

```python
len(df) *0.01 # let us find the top 1 % contributes to 216 records
```

Now cut the top 1 % meaning from 0 to 215
and take rest using iLoc

```python
non_top_1_perc=df.sort_values('price',ascending=False).iloc[216: ]
```

```python
len(non_top_1_perc)
```

```python
sns.scatterplot(x='long',y='lat',data=non_top_1_perc,hue='price')
```

Now it shows the high priced houses are rather scattered based on lat and long

```python
# Let us see if there is a price impact if the house is based on sea or water front
sns.boxplot(x='waterfront',y='price',data=df)
```
